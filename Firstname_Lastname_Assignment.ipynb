Median unit_price:
The median unit_price is 2500.

Is unit_price skewed?
Yes, the unit_price distribution appears to be positively (right) skewed.

Explanation:

->The maximum value (85000) is much higher than most prices.

->The median is only 2500, which is far lower than the highest values.

->A few very expensive items (like laptops) pull the distribution to the right.

This large gap between typical values and the maximum indicates right skewness.
