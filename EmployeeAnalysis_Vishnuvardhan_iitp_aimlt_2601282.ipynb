**Task 1: Create Database**

```python

import sqlite3
import pandas as pd
import random

# Create database
conn = sqlite3.connect('employees.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS employees(
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary REAL,
    years_experience INTEGER,
    performance_score REAL
)
''')

departments = ["Engineering","Sales","Marketing","HR","Finance"]

for i in range(1,41):

    name = f"Employee_{i}"
    department = random.choice(departments)
    salary = random.randint(50000,150000)
    experience = random.randint(1,15)
    performance = round(random.uniform(1.0,5.0),2)

    cursor.execute('''
    INSERT INTO employees(emp_id,name,department,salary,years_experience,performance_score)
    VALUES (?,?,?,?,?,?)
    ''',(i,name,department,salary,experience,performance))

conn.commit()

print("Database and 40 employee records created successfully")
```

```python
df = pd.read_sql_query("SELECT * FROM employees", conn)
df.head()
```

**Task 2: SQL Queries**

```python
#Query 1

query1 = """
SELECT name, department, salary, performance_score
FROM employees
WHERE performance_score >= 4.0
AND years_experience >= 3
ORDER BY performance_score DESC
LIMIT 15
"""

df_query1 = pd.read_sql_query(query1, conn)
df_query1
```

```python
#Query 2

query2 = """
SELECT *
FROM employees
WHERE salary BETWEEN 70000 AND 110000
AND department IN ('Engineering','Sales')
ORDER BY department, salary DESC
"""

df_query2 = pd.read_sql_query(query2, conn)
df_query2
```

```python
#Query 3

query3 = """
SELECT department,
COUNT(*) as employee_count,
AVG(salary) as avg_salary
FROM employees
GROUP BY department
ORDER BY avg_salary DESC
"""

df_query3 = pd.read_sql_query(query3, conn)
df_query3
```

**SQL vs Pandas Comparison**

SQL and Pandas are both powerful tools for data analysis, but they are used in different contexts. SQL is primarily used for querying and managing structured data stored in databases. Pandas, on the other hand, is a Python library used for data manipulation and analysis in memory.

One major difference is the syntax used in filtering and sorting data. In SQL, conditions use operators like = and logical keywords such as AND, while sorting is done using ORDER BY. In Pandas, filtering uses == for equality and & for combining conditions, and sorting is done using the .sort_values() function.

SQL is best used when working directly with relational databases, especially when the dataset is very large and cannot fit into memory. It allows efficient querying without loading the entire dataset into a program. Pandas is more flexible for data analysis, cleaning, visualization, and machine learning tasks because it integrates with the Python ecosystem.

For datasets that are too large to fit into memory, SQL is generally better because database systems are optimized to handle large-scale data storage and querying efficiently without loading everything into RAM.
