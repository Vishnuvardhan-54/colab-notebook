# **Fetching and Analyzing Country Data Using a Public REST API**

### **Task 1 — Fetch and Validate the Response**

```python
# Task 1: Fetch data from REST Countries AP

import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)

    print("Status Code:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print("Total number of countries retrieved:", len(data))
    else:
        print("Error:", response.text)

except Exception as e:
    print("Exception occurred:", str(e))
```

### **Task 2 — Extract and Display Key Fields**

```python
# Task 2: Extract required fields and create DataFrame

import pandas as pd
# Prepare list to store extracted data
country_list = []

for country in data:
    try:
        common_name = country["name"]["common"]
        population = country.get("population", 0)
        region = country.get("region", "Unknown")
        area = country.get("area", 0)

        country_list.append({
            "common_name": common_name,
            "population": population,
            "region": region,
            "area": area
        })

    except KeyError:
        continue

# Create DataFrame
df = pd.DataFrame(country_list)

# Display first 10 rows
df.head(10)
```

### **Task 3 — Summarize the Data**

```python
# Task 3: Data Analysis

# 1. Region with highest total population
region_population = df.groupby("region")["population"].sum()
highest_region = region_population.idxmax()
highest_population = region_population.max()

# 2. Country with largest area
largest_country_row = df.loc[df["area"].idxmax()]

# Print results
print("Region with Highest Total Population:")
print(f"{highest_region} (Population: {highest_population})\n")

print("Country with Largest Area:")
print(f"{largest_country_row['common_name']} (Area: {largest_country_row['area']})")
```
